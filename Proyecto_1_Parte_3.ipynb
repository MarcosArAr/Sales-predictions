## Visualización

#### ¿Cual es el porcentaje de tipos de tiendas respecto al total?

```python
df["Outlet_Type"].value_counts()
```

```python
import matplotlib.pyplot as plt
#Gráfico de torta; cortes ordenados y gráficados contra reloj.
size = df["Outlet_Type"].value_counts()
etiquetas = ['Supermarket Type1', 'Grocery Store', 'Supermarket Type3','Supermarket Type2']
explotar = [0.1, 0, 0, 0] #sirve para destacar alguna porcion. El valor es cuanto se separa de la torta.

fig1 = plt.figure(dpi=100)
ax1 = fig1.add_subplot(1,1,1)

ax1.pie(size, labels=etiquetas, explode=explotar, autopct="%.2f%%", shadow = False, startangle=0)
# autopct = porcentaje automático; shadow = grafico con sombra; startangle = ángulo de inicio
# %.2f%% formato de decimal, viene del lenguaje C. Sí, python está escrito en C.

ax1.axis("equal") #asegura proporcion igualitaria, vale decir, suma de porcentajes 100% (círculo)
plt.title("Outlet Type", fontsize=17)
plt.show()
```

#### ¿Cual es el porcentaje de las distintas categorías de los productos?

```python
df["Item_Type"].value_counts()
```

```python
df["Item_Type"].value_counts().index.unique()
```

```python
import matplotlib.pyplot as plt

#Gráfico de torta; cortes ordenados y gráficados contra reloj.
size = df["Item_Type"].value_counts()
etiquetas = ['Fruits and Vegetables', 'Snack Foods', 'Household', 'Frozen Foods',
       'Dairy', 'Canned', 'Baking Goods', 'Health and Hygiene', 'Soft Drinks',
       'Meat', 'Breads', 'Hard Drinks', 'Others', 'Starchy Foods', 'Breakfast',
       'Seafood']

fig1 = plt.figure(dpi=100)
ax1 = fig1.add_subplot(1,1,1)

ax1.pie(size, labels=etiquetas, autopct="%.2f%%", shadow = False, startangle=40)
# autopct = porcentaje automático; shadow = grafico con sombra; startangle = ángulo de inicio
# %.2f%% formato de decimal, viene del lenguaje C. Sí, python está escrito en C.

ax1.axis("equal") #asegura proporcion igualitaria, vale decir, suma de porcentajes 100% (círculo)
plt.title("Item Type", fontsize=17 )
plt.show()
```

#### ¿Cual es el promedio de ventas de productos segun el tipo de tienda?

```python
df_aux = df.groupby(["Outlet_Type"])[["Item_Outlet_Sales"]].mean().round(2)
df_aux
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.bar(df_aux.index,df_aux["Item_Outlet_Sales"].values, color="black")
plt.xlabel("Outlet Type")
plt.ylabel("Mean Item Outlet Sales")
plt.title("Sales of output items according to store type")
```

#### Salida media del artículo según tipo de artículo de supermercado tipo 3.
Segun los graficos anteriores vemos que el supermercado tipo 3 es el que presenta mayor media de ventas de productos, por ello se analiza el promedio de ventas según categoría del producto respecto al supermercado tipo 3.


**En el grafico vemos que en los supermercados tipo 3 la categoria "Breakfast" presenta mayor ventas respecto a los demas productos y la categoria "Seafood" presenta menor ventas respecto a los demas productos.**

```python
df_grafico = df[df["Outlet_Type"] == "Supermarket Type3"]
df_grafico
```

```python
df_gf = df_grafico.groupby(["Item_Type"])[["Item_Outlet_Sales"]].mean().sort_values(by="Item_Outlet_Sales")
df_gf
```

```python
plt.bar(df_gf.index,df_gf["Item_Outlet_Sales"].values, color="yellowgreen")
plt.xticks(rotation=90)
plt.xlabel("Item type", fontsize=13)
plt.ylabel("Mean Item Outlet Sales", fontsize=13)
plt.title("Mean item outlet according to type of supermarket item type 3", fontsize=15)
```
